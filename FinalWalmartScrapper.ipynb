# Extracting full data

```python
#system libraries
import os
import random
import time
#selenium libraries
from datetime import date
import pyautogui
import webbrowser
# from selenium import webdriver
import pandas as pd
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup, NavigableString, Tag
```

```python
list2=['Auto & Tire', 'Baby', 'Beauty', 'Books', 'Cell Phones', 'Clothing', 'Electronics', 'Food', 'Gifts & Registry', 'Health', 'Home', 'Home Improvement', 'Household Essentials', 'Jewelry', 'Movies', 'Music', 'Office', 'Party & Occasions', 'Patio & Garden', 'Pets', 'Pharmacy', 'Photo Center', 'Sports & Outdoors', 'Toys', 'Video Games']
```

```python
pnames=[]
pprices=[]
pimages=[]
pcategory=[]
failedpages=[]
import pandas as pd
```

```python

for j in range(0,25):
    print("Extracting "+list2[j])
    for i in range(1,26):
#         print(i)
        try:
            
            filename = r'/home/harshit/Desktop/Freelancing/Walmart/Ddata/'+str(list2[j])+'/'+str(list2[j])+str(i)+'.html'
            with open(filename, 'r',encoding='utf8') as f:
                contents = f.read()
                soup = BeautifulSoup(contents)
            temp = soup.find('div',attrs = {'class':'search-product-result'})
            temp = temp.findAll('div',attrs = {'class':'search-result-gridview-item clearfix arrange-fill'})

            for i in range(0,len(temp)):
                
                category= list2[j]
                name=temp[i].find('a',attrs = {'class':'product-title-link line-clamp line-clamp-2 truncate-title'}).text
                pnames.append(name)

                if temp[i].find('span',attrs = {'class':'price price-range'}):
                    p=temp[i].findAll('span',attrs = {'class':'price-group'})
                    price=(p[0].text+' - '+p[1].text)

                elif temp[i].find('span',attrs = {'class':'price-group'}):
                    price=temp[i].find('span',attrs = {'class':'price-group'}).text
                else:
                    price=None

                pprices.append(price)
                image = temp[i].find('img')['data-image-src']
                pimages.append(image)
                pcategory.append(category)
                
        except:
            print("Page "+str(i)+":Failed")
            failedpages.append([(j+1),i]) 

df=pd.DataFrame()
df['Category']=pcategory
df['Product Name']=pnames
df['Price']=pprices
df['Link']=pimages
```
